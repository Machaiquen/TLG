a. Extraer la informaci√≥n del archivo. 

```python
import numpy as np
import pandas as pd

df = pd.read_csv('/content/country_vaccinations.csv')
display(df)
```

b. Mostrar la estructura y tipos de datos de cada columna para identificar qu√© operaciones puedes realizar con cada una de ellas, asegur√°ndote que las columnas con fechas sean del tipo datetime64. 


```python
df['date'] = pd.to_datetime(df['date'])
df.info()
display(df)
```

c. Determinar la cantidad de vacunas aplicadas de cada compa√±√≠a (con base en c√≥mo lo reporta cada pa√≠s en la columna vaccines). 

NOTA: La instrucci√≥n es algo ambigua. Se calcul√≥ la cantidad de vacunas de las empresas por pa√≠s y la cantidad de vacunas por grupo de empresas de vacunas.

```python
vacunas_por_pa√≠s = df[['iso_code', 'vaccines', 'daily_vaccinations']].groupby(['iso_code', 'vaccines']).sum()
vacunas_por_compa√±ia = df[['vaccines', 'daily_vaccinations']].groupby('vaccines').sum()

display(vacunas_por_pa√≠s)
display(vacunas_por_compa√±ia)
```

d. Obtener la cantidad de vacunas aplicadas en todo el mundo. 

```python
n_vacunas_mundial = df['daily_vaccinations'].sum()
print(n_vacunas_mundial)
```

e. Calcular el promedio de vacunas aplicadas por pa√≠s. 

NOTA: Tambi√©n es algo ambigua la instrucci√≥n. Se calcul√≥ promedio de vacunas al d√≠a.

```python
promedio = df[['iso_code', 'daily_vaccinations']].groupby('iso_code').mean()
display(promedio)
```

f. Determinar la cantidad de vacunas aplicadas el d√≠a 29/01/21 en todo el mundo. 



```python
vacunas_dia_29 = df.loc[df['date'] == '2021-01-29', ['daily_vaccinations']].sum()
print(vacunas_dia_29)
```

g. Crear un dataframe nuevo denominado conDiferencias que contenga los datos originales y una columna derivada (diferencias) con las diferencias de aplicaci√≥n entre las columnas daily_vaccionations y daily_vaccionations_raw.  

```python
conDiferencias = df.copy()
conDiferencias['diferencias'] = conDiferencias['daily_vaccinations_raw'] - conDiferencias['daily_vaccinations']
display(conDiferencias)
```

h. Obtener el per√≠odo de tiempo entre el registro con fecha m√°s reciente y el registro con fecha m√°s antigua. 

```python
antigua = df['date'].min()
reciente = df['date'].max()
print(reciente - antigua)
```

i. Crear un dataframe nuevo denominado conCantidad que contenga los datos originales y una columna derivada (canVac) con la cantidad de vacunas utilizadas cada d√≠a (usar la columna vaccines y separar por el car√°cter /). 

```python
conCantidad = df.copy()
conCantidad['canVac'] = conCantidad['vaccines'].str.split('/').str.len()
display(conCantidad)
```

j. Generar un dataframe denominado antes20 con todos los registros que se hayan realizado antes del 20 de diciembre de 2020. 

```python
antes20 = df.copy()
antes20 = antes20[antes20['date'] < '2020-12-20']
display(antes20)
```

k. Obtener un dataframe denominado pfizer con todos los registros donde se haya utilizado 
la vacuna Pfizer.

```python
pfizer = df.copy()
pfizer = pfizer[pfizer['vaccines'].str.contains('Pfizer')]
display(pfizer)
```

l. Almacenar los dataframes generados (conDiferencias,conCantidad,antes20 y pfizer) en 
un archivo de Excel denominado resultadosReto.xlsx, donde cada dataframe ocupe una 
hoja diferente. Se recomienda ver la documentaci√≥n de pd.ExcelWriter

```python
writer = pd.ExcelWriter('resultadosReto.xlsx', engine='xlsxwriter')
conDiferencias.to_excel(writer, sheet_name = 'conDiferencias')
conCantidad.to_excel(writer, sheet_name = 'conCantidad')
antes20.to_excel(writer, sheet_name = 'antes20')
pfizer.to_excel(writer, sheet_name = 'pfizer')
writer.close()
```
